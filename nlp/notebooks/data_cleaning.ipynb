## Unnest nested columns

```python
drop_columns = [
    "place",
    "photos",
    "cashtags",
    "retweet",
    "quote_url",
    "video",
    "near",
    "geo",
    "source",
    "user_rt_id",
    "user_rt",
    "retweet_id",
    "reply_to",
    "retweet_date",
    "translate",
    "trans_src",
    "trans_dest"]

tweets_df.drop(columns=drop_columns, inplace=True)
tweets_df.drop_duplicates(inplace=True)
```

```python
tweets_df["hashtags"].str.replace("'","").str.strip('[]').str.split(',')
```

```python
tweets_df["hashtags"] = tweets_df["hashtags"].str.replace("'","").str.strip('[]').str.split(',')
tweets_df["mentions"] = tweets_df["mentions"].str.replace("'","").str.strip('[]').str.split(',')
tweets_df["urls"] = tweets_df["urls"].str.replace("'","").str.strip('[]').str.split(',')
```

```python
tweets_df["created_at"]
```

```python
pd.to_datetime(tweets_df.created_at,unit='ms')

tweets_df["created_at"] = pd.to_datetime(tweets_df.created_at,unit='ms')
```

```python
tweets_df["hashtags_count"] = tweets_df["hashtags"].apply(lambda x: len(x))
tweets_df["mentions_count"] = tweets_df["mentions"].apply(lambda x: len(x))
```

```python
def unnest(df, tile, explode):
    vals = df[explode].sum(1)
    rs = [len(r) for r in vals]
    a = np.repeat(df[tile].values, rs, axis=0)
    b = np.concatenate(vals.values)
    d = np.column_stack((a, b))
    return pd.DataFrame(d, columns = tile +  explode)
```

```python
df_unnest_hashtags = unnest(tweets_df, list(tweets_df.columns.difference(['hashtags'])),['hashtags'])
df_unnest_hashtags["hashtags"].replace('', np.nan, inplace=True)
df_unnest_hashtags.dropna(how='any', subset=["hashtags"], inplace=True)
df_unnest_hashtags.reset_index(inplace=True)
```

```python
df_unnest_mentions = unnest(tweets_df, list(tweets_df.columns.difference(['mentions'])),['mentions'])
df_unnest_mentions["mentions"].replace('', np.nan, inplace=True)
df_unnest_mentions.dropna(how='any', subset=["mentions"], inplace=True)
df_unnest_mentions.reset_index(inplace=True)
```

```python
def ext_tld(url):
    if not url:
        return [None]*4
    else:
        try:
            path = urlparse(url).path
        except TypeError:
            return [None]*4
        else:
            ext = tldextract.extract(url)
            subdomain = ext.subdomain
            domain = ext.domain
            suffix = ext.suffix
            story = None
            if '-' in path:
                story = "-".join(path.split("-")[:-1]).replace("/","")
            if '@' in path:
                user = path
            
            return subdomain,domain,suffix,story
```

```python
df_unnest_urls = unnest(tweets_df, list(tweets_df.columns.difference(['urls'])),['urls'])
df_unnest_urls["urls"].replace('', np.nan, inplace=True)
df_unnest_urls.dropna(how='any', subset=["urls"], inplace=True)
df_unnest_urls.reset_index(inplace=True)
df_unnest_urls["subdomain"], df_unnest_urls["domain"], df_unnest_urls["suffix"], df_unnest_urls["story"] = zip(*df_unnest_urls["urls"].apply(ext_tld))
df_unnest_urls["story"].replace('', np.nan, inplace=True)
df_unnest_urls.dropna(how='any', subset=["story"], inplace=True)
df_unnest_urls.reset_index(inplace=True)
```

```python
_emojis = [emojis for k in UNICODE_EMOJI.keys() for emojis in UNICODE_EMOJI[k].keys() ]
def get_emoji(text):
    l = re.findall(r'[^\w\s,]', text)
    
    return list(set([e for e in l if e in _emojis]))
```

```python
tweets_df["emojis"] = tweets_df["tweet"].apply(get_emoji)
df_unnest_emojis = unnest(tweets_df, list(tweets_df.columns.difference(['emojis'])),['emojis'])
df_unnest_emojis["emojis"].replace('', np.nan, inplace=True)
df_unnest_emojis.dropna(how='any', subset=["emojis"], inplace=True)
df_unnest_emojis.reset_index(inplace=True)
```

# Data Modeling

```python
l_hashtags = ["created_at","id","user_id","hashtags","likes_count","retweets_count","replies_count"]
l_urls = ["created_at","id","user_id","urls","domain","subdomain","suffix","story","likes_count","retweets_count","replies_count"]
l_mentions = ["created_at","id","user_id","mentions","likes_count","retweets_count","replies_count"]
l_emojis = ["created_at","id","user_id","emojis","likes_count","retweets_count","replies_count"]
l_users = ["created_at","user_id","username","name"]
l_metrics = ["created_at","id","conversation_id","user_id","likes_count","retweets_count","replies_count","hashtags_count","mentions_count"]
```

```python
df_unnest_hashtags = df_unnest_hashtags[l_hashtags]
df_unnest_mentions = df_unnest_mentions[l_mentions]
df_unnest_urls = df_unnest_urls[l_urls]
df_unnest_emojis = df_unnest_emojis[l_emojis]
df_metrics = tweets_df[l_metrics]
df_users = tweets_df[l_users]
```

```python
df_unnest_hashtags = df_unnest_hashtags.reindex(sorted(df_unnest_hashtags.columns), axis=1)
df_unnest_mentions = df_unnest_mentions.reindex(sorted(df_unnest_mentions.columns), axis=1)
df_unnest_urls = df_unnest_urls.reindex(sorted(df_unnest_urls.columns), axis=1)
df_unnest_emojis = df_unnest_emojis.reindex(sorted(df_unnest_emojis.columns), axis=1)
df_metrics = df_metrics.reindex(sorted(df_metrics.columns), axis=1)
df_users = df_users.reindex(sorted(df_users.columns), axis=1)
```

```python
df_unnest_hashtags.to_csv("data/hashtags.csv",index=False)
df_unnest_mentions.to_csv("data/mentions.csv",index=False)
df_unnest_urls.to_csv("data/urls.csv",index=False)
df_unnest_emojis.to_csv("data/emojis.csv",index=False)
df_metrics.to_csv("data/metrics.csv",index=False)
df_users.to_csv("data/users.csv",index=False)
```
